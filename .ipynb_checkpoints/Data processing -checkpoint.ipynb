```python
print('Fraud \n',df.Time[df.Class==1].describe(),'\n',
      '\n Non-Fraud \n',df.Time[df.Class==0].describe())
print('Fraud \n',df.Amount[df.Class==1].describe(),'\n',
      '\n Non-Fraud \n',df.Amount[df.Class==0].describe())
```

```python
plt.figure(figsize=(12,30*4))
import matplotlib.gridspec as gridspec
features = df.iloc[:,0:30].columns
gs = gridspec.GridSpec(30, 1)
for i, feature in enumerate(df[features]):
    ax = plt.subplot(gs[i])
    sns.distplot(df[feature][df.Class == 1], bins=50, color='red')
    sns.distplot(df[feature][df.Class == 0], bins=50, color='blue')
    ax.set_xlabel('')
    ax.set_title('Feature: ' + str(feature))
plt.show()
```
